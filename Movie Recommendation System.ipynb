```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credit=pd.read_csv("tmdb_5000_credits.csv")
```

```python
#MERGING MOVIE AND CREDIT TOGETHER ON THE BASIS OF TITLE
```

```python
movies=movies.merge(credit,on='title')
```

```python
#REMOVING WASTE COLUMNS THAT WE ARE NOT USING FOR THE RECOMMENDATION SYSTEM AND KEEPING ONLY USED COLUMNS
```

```python
movies=movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
#DROPPING NULL VALUES FROM DATASET
movies.dropna(inplace=True)
```

```python
movies.head()
```

```python
#CONVERTING LIST DICTIONARY INTO SIMPLE LIST #GENRES #KEYWORDS AND #CAST
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
        return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies["genres"]
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies["keywords"]
```

```python
import ast
def conv(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i["name"])
            count=count+1
        else:
            break
    return L
```

```python
movies["cast"]=movies["cast"].apply(conv)
```

```python
movies["cast"]
```

```python
movies.head()
```

```python
movies["crew"][0]
```

```python
import ast
def dir(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"]=movies["crew"].apply(dir)
```

```python
movies.crew[0]
```

```python
movies.head()
```

```python
#Now Converting Overview Into List
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies.head()
```

```python
#Now Merging overview,genres,keywords, cast And crew In 1 Single Column Named As Tag
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
movies["tags"]
```

```python
mrs=movies[["movie_id","title","tags"]]
```

```python
mrs
```

```python
mrs["tags"]=mrs["tags"].fillna("No_Movies")
```

```python
mrs["tags"] = mrs["tags"].apply(lambda x: " ".join(x))
```

```python
mrs["tags"] = mrs["tags"].apply(lambda x:x.lower())
```

```python
mrs
```

```python
mrs.head()
```

```python
#Skilearn is a module pack of CountVectorizer Package
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000, stop_words="english")
vect=cv.fit_transform(mrs["tags"]).toarray()
cv.get_feature_names_out()
```

```python
#nltk natural language toolkit
#stem used to Give Only 1 Meaning Word And Removes Remaining Synonyms
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
ps.stem("Loving")
```

```python
mrs
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
mrs["tags"]=mrs["tags"].apply(stem)
```

```python
mrs
```

```python
#Finding The Angles Between 2 movies for that-:
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar=cosine_similarity(vect)
```

```python
similar
```

```python
sorted([similar],reverse=True)
```

```python
list(enumerate(similar[0]))
```

```python
def recommend(movie):
    movie_index=mrs[mrs["title"]==movie].index[0]
    distance=similar[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(mrs.iloc[i[0]].title)
```

```python
sorted(list(enumerate(similar[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
recommend("Iron Man")
```
